```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import sklearn
import pandas as pd
import matplotlib
import seaborn as sns

data=pd.read_csv("SampleDatasets/circles0.3.csv")

# Use the 'hue' argument to provide a factor variable
sns.lmplot(x="x1", y="x2", data=data, fit_reg=False, height=5, aspect=1.2, hue='label', legend=False, markers=["+", "^"])

# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
plt.title(f"Circles data")
plt.xlabel("X")
plt.ylabel("Y");
```
